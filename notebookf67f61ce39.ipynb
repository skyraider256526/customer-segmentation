## Removing null and duplicates

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.describe()
data.info()
```

```python
#Removing CustomerID
X = pd.get_dummies( data.drop(columns=['CustomerID','InvoiceNo','StockCode','Description','InvoiceDate']))
X
```

```python
kmeans = KMeans(init='k-means++', max_iter=300, n_init=10,random_state=4)
```

```python
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(X)
```

```python
visualizer = kelbow_visualizer(kmeans, scaled_features, k=(2,10))
```

```python
visualizer.elbow_value_
```

```python
kmeans = KMeans(init='k-means++', max_iter=300, n_init=10,random_state=4,n_clusters=visualizer.elbow_value_ )
```

```python
y_means = kmeans.fit_predict(scaled_features)
```

```python
data['Cluster'] = y_means
```
